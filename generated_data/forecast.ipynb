```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('generated_data/forecast_cheikh.xlsx', sheet_name=0)
```

```python
data = data.drop(data.columns[[0, 1]], axis=1)
```

```python
data.head()
```

```python
data.columns = data.iloc[0]
data = data.drop(data.index[0]).reset_index(drop=True)
data.head()
```

```python
# Remove rows where all week columns are empty (NaN or blank)
week_columns = [col for col in data.columns if col.startswith('W')]
data = data.dropna(subset=week_columns, how='all')
data
```

```python
data = data.fillna(0)
data
```
